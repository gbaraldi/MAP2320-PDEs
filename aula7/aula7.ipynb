```python
import numpy as np
from matplotlib import pyplot as plt
```

A solução tem é a soma da solução permanente e do transitório

```python
def soluchom(cs,ns,x,t):
    sine = np.sin(ns*np.pi*x/L)*np.exp(-(ns**2*np.pi**2/L**2)*t)
    val = np.sum((sine*cs)) # parte transitória
    val = T1 + (T2 - T1)/L *x + val #Parte permanente
    return val
```

```python
L = 100
T1 = 10
T2 = 20
Tm = 80
```

f(x) = T1 0<L/3 
Tm L/3->2L/3
T2 2L/3 -> L

São 3 constantes, portanto a solução vai ser a soma das 3 constantes

```python
# Primeira parte constante de f(x)
def calculacs1(ns):
    c = 0
    d = 1/3
    cn = (-2*(T1)/(np.pi*ns))*(np.cos(np.pi*ns*d)-np.cos(np.pi*ns*c))
    return cn
```

```python
# Segunda parte constante de f(x)
def calculacs2(ns):
    c = 1/3
    d = 2/3
    cn = (-2*(Tm)/(np.pi*ns))*(np.cos(np.pi*ns*d)-np.cos(np.pi*ns*c))
    return cn
```

```python
# Terceira parte constante de f(x)
def calculacs3(ns):
    c = 2/3
    d = 1
    cn = (-2*(T2)/(np.pi*ns))*(np.cos(np.pi*ns*d)-np.cos(np.pi*ns*c))
    return cn
```

Além disso temos que subtrair a solução permanente

```python
# Parte constante da solução permanente
def calculacs4(ns):
    c = 0
    d = 1
    cn = (-2*(T1)/(np.pi*ns))*(np.cos(np.pi*ns*d)-np.cos(np.pi*ns*c))
    return cn
```

```python
def calculacs5(ns):
    c = 0
    d = 1
    s1 = np.pi*ns*d
    s2 = np.pi*ns*c
    cn = (-2*(L*0.1)/(np.pi**2*ns**2))*(-(s1)*np.cos(s1) +(s2)*np.cos(s2) + np.sin(s1) - np.sin(s2))
    return cn
```

```python
xs = np.linspace(0,L,num=100)
```

```python
ts = np.geomspace(1,10000, num= 9)
ts[0] = 0
```

```python
n = 100
ns = np.arange(1,n)
```

```python
fig, axes = plt.subplots(3,3, figsize = (18,9))
k = 0
for i in range(0,3):
    for j in range(0,3):
        axes[i,j].plot(xs,[soluchom(calculacs1(ns) + calculacs2(ns) + calculacs3(ns) + calculacs4(ns) + calculacs5(ns),ns,x,ts[k]) for x in xs])
        axes[i,j].set_title(repr(ts[k])[0:5])
        k +=1
```
