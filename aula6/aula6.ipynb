```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
L = 40
```

```python
n = 8
ns = np.arange(1,n*2,step = 2)
```

```python
def soluchom(cs,ns,x,t):
    sine = np.sin(ns*np.pi*x/L)*np.exp(-(ns**2*np.pi**2/L**2)*t)
    val = np.sum((sine*cs))
    return val
```

```python
def calculacs(ns):
    cn = 4*L/np.pi**2 * np.where(ns%4 == 1, 1,-1) /(ns**2)
    return cn
```

```python
xs = np.linspace(0,40,num=100)
```

```python
ts = np.geomspace(1,1280, num= 9)
ts[0] = 0
```

```python
repr(ts[1])[0:5]
```

```python
fig, axes = plt.subplots(3,3, figsize = (18,9))
k = 0
for i in range(0,3):
    for j in range(0,3):
        axes[i,j].plot(xs,[soluchom(calculacs(ns),ns,x,ts[k]) for x in xs])
        axes[i,j].set_title(repr(ts[k])[0:5])
        axes[i,j].set_ylim(0,20)
        k +=1
```

```python
def calculacs2(ns):
    c = 1/4
    d = 3/4
    cn = (-2*(L/2)/(np.pi*ns))*(np.cos(np.pi*ns*d)-np.cos(np.pi*ns*c))
    return cn
```

```python
n = 50
ns = np.arange(1,n*2,step = 2)
```

```python
fig, axes = plt.subplots(3,3, figsize = (18,9))
k = 0
for i in range(0,3):
    for j in range(0,3):
        axes[i,j].plot(xs,[soluchom(calculacs2(ns),ns,x,ts[k]) for x in xs])
        axes[i,j].set_title(repr(ts[k])[0:5])
        axes[i,j].set_ylim(0,25)
        k +=1
```
