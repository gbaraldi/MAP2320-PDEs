```python
import numpy as np
from matplotlib import pyplot as plt
```

cn = 2/L integral de f(x)e^x * sen(n*pi*x/L)

f(x) = L/2 de L/4 até 3/4L

Como f(x) é constante a função pode ser transformada em:

cn = 2/L * L/2 integral de e^x * sen(n*pi*x/l)

A integral indefinida dá:

$e^x(sin(pi*n*x) - pi*n*cos(pi*n*x))/(pi^2*n^2 + 1)$

```python
L = 1
```

```python
def indefinida(x,n):
    return np.e**x * (np.sin(np.pi*n*x) - np.pi*n*np.cos(np.pi*n*x))/(np.pi**2*n**2 + 1)
```

```python
def cn(n):
    return indefinida(3/4,n) - indefinida(1/4,n)
```

```python
cn(2)
```

```python
n = 100
ns = np.arange(1,n)
```

```python
ts = np.linspace(0,0.2, num= 9)
ts[0] = 0
```

```python
xs = np.linspace(0,L,num=100)
```

```python
def soluchom(cs,ns,x,t):
    sine = np.sin(ns*np.pi*x/L)*np.exp(-(ns**2*np.pi**2/L**2)*t)
    val = np.sum((sine*cs*np.exp(-x-t))) # parte transitóriaParte permanente
    return val
```

```python
fig, axes = plt.subplots(3,3, figsize = (18,9))
k = 0
for i in range(0,3):
    for j in range(0,3):
        axes[i,j].plot(xs,[soluchom(cn(ns),ns,x,ts[k]) for x in xs])
        axes[i,j].set_title(repr(ts[k])[0:5])
        axes[i,j].set_ylim(0,1)
        k +=1
```
