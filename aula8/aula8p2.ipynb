```python
import numpy as np
from matplotlib import pyplot as plt
```

Condições de Neumman portanto série de cossenos

A f(x) é uma reta da forma 3x/2

Esta é uma função tabelada simples portanto é só olhar nela.

```python
L = 40
```

```python
def calculacn(ns):
    c = 0
    d = 1
    s1 = np.pi*ns*d
    s2 = np.pi*ns*c
    cn = (2*(L)/(np.pi**2*ns**2))*((s1)*np.sin(s1) - (s2)*np.sin(s2) + np.cos(s1) - np.cos(s2))
    return cn
```

```python
def calculacn0():
    c = 0
    d = 1
    return L/2*(d**2 - c**2)
```

```python
def calculacns(ns):
    cn = calculacn(ns[1:])
    cn = np.insert(cn,0,calculacn0())
    return cn
```

```python
def soluchom(cs,ns,x,t):
    sine = np.cos(ns*np.pi*x/L)*np.exp(-(ns**2*np.pi**2/L**2)*t)
    val = np.sum((sine*cs)) # parte transitóriarmanente
    return val
```

```python
n = 100
ns = np.arange(0,n)
```

```python
ts = np.geomspace(1,1000, num= 9)
ts[0] = 0
```

```python
xs = np.linspace(0,L,num=100)
```

```python
fig, axes = plt.subplots(3,3, figsize = (18,9))
k = 0
for i in range(0,3):
    for j in range(0,3):
        axes[i,j].plot(xs,[soluchom(calculacns(ns),ns,x,ts[k]) for x in xs])
        axes[i,j].set_title(repr(ts[k])[0:5])
        axes[i,j].set_ylim(0,40)
        k +=1
```

O valor final é 20, o que faz sentido, pois 20 é o valor médio da f(x) da temperatura inicial
